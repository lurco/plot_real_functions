```python
def f(x):
    return x ** 3 + x ** 2 - 2 * x - 1
```

```python
# plot f(x)
import numpy as np
import matplotlib.pyplot as plt
fig, ax = plt.subplots()
x = np.linspace(-3, 3, 100)
y = f(x)
y2 = f(-x)
y3 = -f(x)
ax.spines['left'].set_position('zero')  # Y-axis at x=0
ax.spines['bottom'].set_position('zero')  # X-axis at y=0
ax.plot(x, y)
ax.plot(x, y2)
ax.set_xlabel("x", fontsize=12, labelpad=15)
ax.set_ylabel("y", fontsize=12, labelpad=15, rotation=0)
ax.yaxis.set_label_coords(0.55, 0.95)
ax.xaxis.set_label_coords(0.95, .5)
# plt.plot(x, y3)
plt.show()
```
